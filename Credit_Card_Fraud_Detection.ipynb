```python
# Separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraudulent = credit_card_data[credit_card_data.Class == 1]
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample,fraudulent],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns = 'Class',axis =1)
```

```python
Y = new_dataset['Class']
```

```python
X.head()
```

```python
Y
```

Split the data into Training data & Testing Data

```python
X_train , X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2,stratify = Y)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy score

```python
# Training data accuracy
training_data_accuracy = model.predict(X_train)
```

```python
print("Training Accuracy Score: ",accuracy_score(Y_train,training_data_accuracy))
```

```python
#Testing data accuracy
testing_data_accuracy = model.predict(X_test)
print("Testing accuracy score: ",accuracy_score(Y_test, testing_data_accuracy))
```

Model Training - XGBoost

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train,Y_train)
```

Model Evaluation - XGBoost 

Accuracy score

```python
# Training data accuracy
training_data_accuracy = classifier.predict(X_train)
print("Training Accuracy Score: ",accuracy_score(Y_train,training_data_accuracy))
```

```python
#Testing data accuracy
testing_data_accuracy = classifier.predict(X_test)
print("Testing accuracy score: ",accuracy_score(Y_test, testing_data_accuracy))
```
